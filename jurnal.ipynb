```python
import matplotlib.pyplot as plt

# Data
data = [
    [1, 2, -1, 1],
    [0, 2, 1, 2],
    [2, 1, 1, 0],
    [1, 0, -1, 0]
]

# Flatten the data into a single list
flat_data = [item for sublist in data for item in sublist]

# Plot histogram
plt.hist(flat_data, bins=range(min(flat_data), max(flat_data) + 2, 1), edgecolor='black')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Histogram Data')
plt.grid(True)
plt.show()
```
